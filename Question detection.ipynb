```python
import re
import string
from IPython.display import display

# -------------------------------
# 1. Cleaning function
# -------------------------------
def clean_basic(text: str) -> str:
    text = str(text)

    # remove HTML tags
    text = re.sub(r"<.*?>", " ", text)

    # remove [math] ... [/math] tags
    text = re.sub(r"\[/?math\]", " ", text)

    # remove punctuation
    text = text.translate(str.maketrans("", "", string.punctuation))

    # collapse extra whitespace
    text = re.sub(r"\s+", " ", text).strip()

    return text

# -------------------------------
# 2. Apply cleaning to both columns
# -------------------------------
df["question1_clean"] = df["question1"].apply(clean_basic)
df["question2_clean"] = df["question2"].apply(clean_basic)

# -------------------------------
# 3. Preview
# -------------------------------
display(df[["question1", "question2", "question1_clean", "question2_clean"]].head(10))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
import numpy as np
import pandas as pd

# -----------------------------
# 1. Build TF-IDF on training data (Q1 + Q2 together)
# -----------------------------
# We do NOT modify train_clean / test_clean; we only read from them.

# Combine all questions from training set to build the vocabulary
train_questions_all = pd.concat(
    [train_clean["question1_clean"], train_clean["question2_clean"]],
    axis=0
)

tfidf = TfidfVectorizer(
    lowercase=True,
    ngram_range=(1, 2),     # unigrams + bigrams (you can change this)
    min_df=2,               # ignore very rare terms (optional)
    max_features=100000     # cap features (optional)
)

tfidf.fit(train_questions_all)

# -----------------------------
# 2. Transform Q1 and Q2 separately with SAME TF-IDF
# -----------------------------
# Train
q1_train_tfidf = tfidf.transform(train_clean["question1_clean"])
q2_train_tfidf = tfidf.transform(train_clean["question2_clean"])

# Test
q1_test_tfidf = tfidf.transform(test_clean["question1_clean"])
q2_test_tfidf = tfidf.transform(test_clean["question2_clean"])

# -----------------------------
# 3. Row-wise cosine similarity between Q1 and Q2
#    (TF-IDF vectors are L2-normalized by default → dot product = cosine)
# -----------------------------
# Element-wise multiply and sum across features to get similarity per row
train_sim = q1_train_tfidf.multiply(q2_train_tfidf).sum(axis=1)
test_sim  = q1_test_tfidf.multiply(q2_test_tfidf).sum(axis=1)

# Convert from matrix to 1D numpy array
X_train = np.asarray(train_sim).ravel().reshape(-1, 1)   # shape: (n_samples, 1)
X_test  = np.asarray(test_sim).ravel().reshape(-1, 1)

# -----------------------------
# 4. Prepare labels (WITHOUT changing the original DataFrame)
# -----------------------------
y_train = train_clean["is_duplicate"].astype(int).values
y_test  = test_clean["is_duplicate"].astype(int).values

# -----------------------------
# 5. Train a simple classifier on cosine similarity
# -----------------------------
clf = LogisticRegression()
clf.fit(X_train, y_train)

# -----------------------------
# 6. Evaluate on test set
# -----------------------------
y_pred = clf.predict(X_test)

print("🔢 Test Accuracy:", accuracy_score(y_test, y_pred))
print("\n📊 Classification Report:\n")
print(classification_report(y_test, y_pred))
```

```python
# If gensim is not installed, uncomment this:
# !pip install gensim

from gensim.models import Word2Vec
import numpy as np
import pandas as pd
from sklearn.metrics import f1_score, accuracy_score, classification_report

# ---------------------------------------------------
# 1. Prepare data for Word2Vec training (from TRAIN only)
# ---------------------------------------------------

# Use question1_clean and question2_clean text from train_clean
train_q1_texts = train_clean["question1_clean"].astype(str).tolist()
train_q2_texts = train_clean["question2_clean"].astype(str).tolist()

# Tokenize by simple whitespace split (assuming text already cleaned)
sentences = [q.split() for q in train_q1_texts] + [q.split() for q in train_q2_texts]

# ---------------------------------------------------
# 2. Train Word2Vec model
# ---------------------------------------------------
w2v_model = Word2Vec(
    sentences=sentences,
    vector_size=300,   # embedding size
    window=5,
    min_count=2,
    workers=4,
    sg=1               # 1 = skip-gram, 0 = CBOW
)

word_vectors = w2v_model.wv
embed_dim = word_vectors.vector_size

# ---------------------------------------------------
# 3. Helper: get sentence vector (average of word vectors)
# ---------------------------------------------------
def sentence_to_vec(sentence, word_vectors, embed_dim):
    tokens = sentence.split()
    vecs = []
    for tok in tokens:
        if tok in word_vectors:
            vecs.append(word_vectors[tok])
    if not vecs:  # no known tokens
        return np.zeros(embed_dim, dtype=np.float32)
    vecs = np.array(vecs)
    return vecs.mean(axis=0)

def batch_sentence_vectors(series, word_vectors, embed_dim):
    return np.vstack([
        sentence_to_vec(text, word_vectors, embed_dim)
        for text in series.astype(str)
    ])

# ---------------------------------------------------
# 4. Build sentence vectors for Q1 and Q2 (train + test)
# ---------------------------------------------------

# Train vectors
q1_train_vec = batch_sentence_vectors(train_clean["question1_clean"], word_vectors, embed_dim)
q2_train_vec = batch_sentence_vectors(train_clean["question2_clean"], word_vectors, embed_dim)

# Test vectors
q1_test_vec = batch_sentence_vectors(test_clean["question1_clean"], word_vectors, embed_dim)
q2_test_vec = batch_sentence_vectors(test_clean["question2_clean"], word_vectors, embed_dim)

# ---------------------------------------------------
# 5. Cosine similarity between Q1 and Q2 vectors
# ---------------------------------------------------
def cosine_sim_batch(a, b, eps=1e-8):
    # a, b: shape (n_samples, embed_dim)
    dot_product = np.sum(a * b, axis=1)
    norm_a = np.linalg.norm(a, axis=1)
    norm_b = np.linalg.norm(b, axis=1)
    return dot_product / (norm_a * norm_b + eps)

train_sim = cosine_sim_batch(q1_train_vec, q2_train_vec)
test_sim  = cosine_sim_batch(q1_test_vec, q2_test_vec)

# ---------------------------------------------------
# 6. Prepare labels (without modifying DataFrames)
# ---------------------------------------------------
y_train = train_clean["is_duplicate"].astype(int).values
y_test  = test_clean["is_duplicate"].astype(int).values

# ---------------------------------------------------
# 7. Find best similarity threshold on TRAIN set
# ---------------------------------------------------
thresholds = np.linspace(-1.0, 1.0, 201)  # cosine can be [-1, 1]
best_threshold = None
best_f1 = -1.0

for thr in thresholds:
    y_pred_train = (train_sim >= thr).astype(int)
    f1 = f1_score(y_train, y_pred_train)
    if f1 > best_f1:
        best_f1 = f1
        best_threshold = thr

print(f"🔍 Best threshold on TRAIN: {best_threshold:.4f} with F1 = {best_f1:.4f}")

# ---------------------------------------------------
# 8. Evaluate on TEST set using best threshold
# ---------------------------------------------------
y_pred_test = (test_sim >= best_threshold).astype(int)

test_acc = accuracy_score(y_test, y_pred_test)
print(f"\n✅ Test Accuracy (Word2Vec + cosine threshold): {test_acc:.4f}\n")

print("📊 Classification Report (Test):\n")
print(classification_report(y_test, y_pred_test))
```

```python
# Siamese LSTM on question1_clean & question2_clean
# Assumes: train_clean, test_clean already exist from train_test_split

import numpy as np
import pandas as pd
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, LSTM, GRU, Bidirectional
from tensorflow.keras.layers import Dense, Dropout, Lambda, Concatenate
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import AUC
import tensorflow as tf

# ---------------------------------------------------
# 1. Prepare texts and labels (WITHOUT changing DataFrames)
# ---------------------------------------------------

# Use only the cleaned text columns
train_q1 = train_clean["question1_clean"].astype(str).values
train_q2 = train_clean["question2_clean"].astype(str).values
test_q1  = test_clean["question1_clean"].astype(str).values
test_q2  = test_clean["question2_clean"].astype(str).values

# Labels
y_train = train_clean["is_duplicate"].astype(int).values
y_test  = test_clean["is_duplicate"].astype(int).values

# ---------------------------------------------------
# 2. Tokenizer on TRAIN text only (q1 + q2)
# ---------------------------------------------------
all_train_text = np.concatenate([train_q1, train_q2])

max_vocab_size = 50000   # you can adjust
tokenizer = Tokenizer(num_words=max_vocab_size, oov_token="<OOV>")
tokenizer.fit_on_texts(all_train_text)

# Convert to sequences
X_train_q1_seq = tokenizer.texts_to_sequences(train_q1)
X_train_q2_seq = tokenizer.texts_to_sequences(train_q2)
X_test_q1_seq  = tokenizer.texts_to_sequences(test_q1)
X_test_q2_seq  = tokenizer.texts_to_sequences(test_q2)

# Padding
max_len = 40  # you can tune this (e.g., 40/50/60)
X_train_q1 = pad_sequences(X_train_q1_seq, maxlen=max_len, padding="post", truncating="post")
X_train_q2 = pad_sequences(X_train_q2_seq, maxlen=max_len, padding="post", truncating="post")
X_test_q1  = pad_sequences(X_test_q1_seq,  maxlen=max_len, padding="post", truncating="post")
X_test_q2  = pad_sequences(X_test_q2_seq,  maxlen=max_len, padding="post", truncating="post")

vocab_size = min(max_vocab_size, len(tokenizer.word_index) + 1)

# ---------------------------------------------------
# 3. Build Siamese LSTM / GRU encoder
# ---------------------------------------------------

embed_dim = 128
lstm_units = 64

# Shared input shapes
input_q1 = Input(shape=(max_len,), name="input_q1")
input_q2 = Input(shape=(max_len,), name="input_q2")

# Shared embedding layer
embedding_layer = Embedding(
    input_dim=vocab_size,
    output_dim=embed_dim,
    input_length=max_len,
    mask_zero=True,
    name="shared_embedding"
)

# Shared encoder: you can switch LSTM <-> GRU easily
# encoder = Bidirectional(LSTM(lstm_units, return_sequences=False, dropout=0.2, recurrent_dropout=0.2),
#                         name="shared_bi_lstm")
encoder = Bidirectional(GRU(lstm_units, return_sequences=False, dropout=0.2, recurrent_dropout=0.2),
                        name="shared_bi_gru")

# Encode both questions with shared weights
x1 = embedding_layer(input_q1)
x2 = embedding_layer(input_q2)

enc_q1 = encoder(x1)
enc_q2 = encoder(x2)

# ---------------------------------------------------
# 4. Similarity / comparison layer
# ---------------------------------------------------
# L1 distance (absolute difference)
l1_dist = Lambda(lambda tensors: tf.math.abs(tensors[0] - tensors[1]),
                 name="l1_distance")([enc_q1, enc_q2])

# Element-wise square difference (L2-style)
l2_dist = Lambda(lambda tensors: tf.math.square(tensors[0] - tensors[1]),
                 name="l2_distance")([enc_q1, enc_q2])

# Concatenate original encodings + distances
merged = Concatenate(name="merge_features")([enc_q1, enc_q2, l1_dist, l2_dist])

# ---------------------------------------------------
# 5. Classifier head (sigmoid output)
# ---------------------------------------------------
dense = Dense(128, activation="relu")(merged)
dense = Dropout(0.3)(dense)
dense = Dense(64, activation="relu")(dense)
dense = Dropout(0.3)(dense)
output = Dense(1, activation="sigmoid", name="output")(dense)

model = Model(inputs=[input_q1, input_q2], outputs=output)
model.compile(
    loss="binary_crossentropy",
    optimizer=Adam(learning_rate=1e-3),
    metrics=["accuracy", AUC(name="auc")]
)

model.summary()

# ---------------------------------------------------
# 6. Train the Siamese model
# ---------------------------------------------------
history = model.fit(
    [X_train_q1, X_train_q2],
    y_train,
    batch_size=256,
    epochs=5,              # increase for better performance
    validation_split=0.1,  # from training data only
    verbose=1
)

# ---------------------------------------------------
# 7. Evaluate on TEST set
# ---------------------------------------------------
test_preds_proba = model.predict([X_test_q1, X_test_q2])
test_preds = (test_preds_proba >= 0.5).astype(int).ravel()

from sklearn.metrics import accuracy_score, classification_report

print("✅ Test Accuracy (Siamese RNN):", accuracy_score(y_test, test_preds))
print("\n📊 Classification Report (Test):\n")
print(classification_report(y_test, test_preds))
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# ---------------------------------------------------
# 1. Read data from existing DataFrames (NO modification)
# ---------------------------------------------------
train_q1 = train_clean["question1_clean"].astype(str).tolist()
train_q2 = train_clean["question2_clean"].astype(str).tolist()
test_q1  = test_clean["question1_clean"].astype(str).tolist()
test_q2  = test_clean["question2_clean"].astype(str).tolist()

y_train = train_clean["is_duplicate"].astype(int).values
y_test  = test_clean["is_duplicate"].astype(int).values

# ---------------------------------------------------
# 2. Load Sentence-BERT (Bi-Encoder)
#    You can change the model name if needed
# ---------------------------------------------------
model = SentenceTransformer("paraphrase-MiniLM-L6-v2")

# ---------------------------------------------------
# 3. Encode Q1 and Q2 into dense sentence embeddings
# ---------------------------------------------------
emb_train_q1 = model.encode(train_q1, batch_size=64, show_progress_bar=True)
emb_train_q2 = model.encode(train_q2, batch_size=64, show_progress_bar=True)

emb_test_q1  = model.encode(test_q1,  batch_size=64, show_progress_bar=True)
emb_test_q2  = model.encode(test_q2,  batch_size=64, show_progress_bar=True)

# Each emb_* has shape: (n_samples, embedding_dim)

# ---------------------------------------------------
# 4. Build features for classifier from embeddings
#    Common trick: [|u - v| , u * v]  (element-wise)
# ---------------------------------------------------
def build_features(e1, e2):
    # e1, e2: (n_samples, d)
    abs_diff = np.abs(e1 - e2)
    prod     = e1 * e2
    return np.concatenate([abs_diff, prod], axis=1)

X_train = build_features(emb_train_q1, emb_train_q2)
X_test  = build_features(emb_test_q1,  emb_test_q2)

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)

# ---------------------------------------------------
# 5. Train a simple classifier (Logistic Regression)
# ---------------------------------------------------
clf = LogisticRegression(
    max_iter=1000,
    n_jobs=-1,
    verbose=1
)
clf.fit(X_train, y_train)

# ---------------------------------------------------
# 6. Evaluate on TEST set
# ---------------------------------------------------
y_pred = clf.predict(X_test)
```

```python
# ---------------------------------------------------
# 0. Install dependencies (run once)
# ---------------------------------------------------
# If not already installed:
# !pip install -U sentence-transformers

from sentence_transformers import CrossEncoder, InputExample
from torch.utils.data import DataLoader
import numpy as np
from sklearn.metrics import accuracy_score, classification_report, f1_score

# ---------------------------------------------------
# 1. Prepare data (READ-ONLY from train_clean / test_clean)
# ---------------------------------------------------

# Text columns
train_q1 = train_clean["question1_clean"].astype(str).tolist()
train_q2 = train_clean["question2_clean"].astype(str).tolist()
test_q1  = test_clean["question1_clean"].astype(str).tolist()
test_q2  = test_clean["question2_clean"].astype(str).tolist()

# Labels
y_train = train_clean["is_duplicate"].astype(float).values  # float for regression-style score
y_test  = test_clean["is_duplicate"].astype(int).values     # keep int for metrics

print("Train size:", len(train_q1))
print("Test size:", len(test_q1))

# ---------------------------------------------------
# 2. Build training examples for Cross-Encoder
# ---------------------------------------------------

train_examples = [
    InputExample(texts=[q1, q2], label=label)
    for q1, q2, label in zip(train_q1, train_q2, y_train)
]

# Optional: you can also create a small validation set from train_clean if you want

train_dataloader = DataLoader(train_examples, batch_size=16, shuffle=True)

# ---------------------------------------------------
# 3. Initialize Cross-Encoder
# ---------------------------------------------------
# This model is already pre-trained on Quora-style duplicates, but we still fine-tune on your data.
# You can change to another cross-encoder model if needed.

model_name = "cross-encoder/quora-roberta-base"
cross_encoder = CrossEncoder(model_name, num_labels=1)  # regression-style output in [0,1]

# ---------------------------------------------------
# 4. Train Cross-Encoder
# ---------------------------------------------------

num_epochs = 2  # start with 1; you can increase to 2-3 if Colab can handle it
warmup_steps = int(0.1 * len(train_dataloader))  # 10% of steps for warmup

cross_encoder.fit(
    train_dataloader=train_dataloader,
    epochs=num_epochs,
    warmup_steps=warmup_steps,
    output_path="output/cross_encoder_quora",  # saves model here (optional)
    use_amp=True  # mixed precision for speed on GPU
)

# If you want to reload later:
# cross_encoder = CrossEncoder("output/cross_encoder_quora", num_labels=1)

# ---------------------------------------------------
# 5. Predict scores on TRAIN + TEST (for threshold tuning)
# ---------------------------------------------------

# CrossEncoder expects list of [q1, q2] pairs
train_pairs = list(zip(train_q1, train_q2))
test_pairs  = list(zip(test_q1, test_q2))

# Predict similarity scores in [0, 1]
train_scores = cross_encoder.predict(train_pairs, batch_size=32)
test_scores  = cross_encoder.predict(test_pairs,  batch_size=32)

print("Example train scores:", train_scores[:5])
print("Example test scores:",  test_scores[:5])

# ---------------------------------------------------
# 6. Find best threshold on TRAIN (based on F1)
# ---------------------------------------------------

thresholds = np.linspace(0.0, 1.0, 101)
best_thr = 0.5
best_f1 = -1.0

for thr in thresholds:
    y_train_pred = (train_scores >= thr).astype(int)
    f1 = f1_score(y_train.astype(int), y_train_pred)
    if f1 > best_f1:
        best_f1 = f1
        best_thr = thr

print(f"🔍 Best threshold on TRAIN: {best_thr:.3f} with F1 = {best_f1:.4f}")

# ---------------------------------------------------
# 7. Evaluate on TEST set using best threshold
# ---------------------------------------------------

y_test_pred = (test_scores >= best_thr).astype(int)

test_acc = accuracy_score(y_test, y_test_pred)
print(f"\n✅ Test Accuracy (Cross-Encoder): {test_acc:.4f}\n")

print("📊 Classification Report (Test):\n")
print(classification_report(y_test, y_test_pred))
```

```python
def compare_questions(q1: str, q2: str) -> str:
    q1 = (q1 or "").strip()
    q2 = (q2 or "").strip()

    if not q1 or not q2:
        return "Please enter both questions."

    # CrossEncoder expects a list of [q1, q2] pairs
    score = float(model.predict([[q1, q2]])[0])
    label = "Duplicate" if score >= THRESHOLD else "Not Duplicate"

    # Only return the label as you requested
    return label
```
